```python
#belli bir aralıktaki asal sayıları yazdırma

a=int(input(""" '1' sayısı asal sayı olarak kabul edilmez çünkü her sayıya tam bölünür..
            Bu yüzden '2' den başlayacak şekilde son sayıyı seçerek asallık sorgusu yapılması için bir aralık belirleyiniz.."""))

for i in range(2,a+1):
    notprime=0
    for j in range(2,(i//2)+1):
        if i%j==0:
            notprime=1
            break
    if notprime==0:
        print("{} sayısı bir asal sayıdır".format(i))
            
```
