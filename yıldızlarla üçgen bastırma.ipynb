```python
#yıldızlarla üçgen bastırma(1)

a=int(input("lütfen üçgen limitini, yani kaç satırlık üçgen olacağını belirlemek için sayı giriniz.."))

for i in range(1,a+1):
    for j in range(1,i+1):
        print("*",end=" ")
    print("\n")
```

```python
#yıldızlarla üçgen bastırma(2)

a=int(input("lütfen üçgen limitini, yani kaç satırlık üçgen olacağını belirlemek için sayı giriniz.."))

for i in range(1,a+1):
    for j in range((a+1)-i,0,-1):
        print("*",end=" ")
    print("\n")
```

```python
#yıldızlarla üçgen bastırma(3)

a=int(input("lütfen üçgen limitini, yani kaç satırlık üçgen olacağını belirlemek için sayı giriniz.."))

for i in range(1,2*a+1,2):
    for j in range(2*a-i,0,-1):
        print("*",end=" ")
        
    print("\n")    
            
```
