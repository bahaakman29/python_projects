```python
#example of oop-inheritance

class Insan():
    
    def __init__(self,isim,soyisim):
        self.isim=isim
        self.soyisim=soyisim
    def bilgileriYazdir(self):
        print("isim: {} ve soyisim: {}".format(self.isim,self.soyisim))

class Ogrenci(Insan):
    
    ogrenciListesi=[]
    
    def __init__(self,isim,soyisim,ogrenciNo,ogrenciNot,ders,ogrenciGecmeKalma=0):
        super().__init__(isim,soyisim)
        self.ogrenciNo=ogrenciNo
        self.ogrenciNot=ogrenciNot
        self.ders=ders
        self.ogrenciGecmeKalma=ogrenciGecmeKalma
        Ogrenci.ogrenciListesi.append(self)
    
    def bilgileriYazdir(self):
        super().bilgileriYazdir(self)
        print("ogrenci no:{} ogrenci notu: {} ".format(self.ogrenciNo,self.ogrenciNot))

class Ogretmen(Insan):
    def __init__(self,isim,soyisim,departman):
        super().__init__(isim,soyisim)
        self.departman=departman
    def bilgileriYazdir(self):
        super().bilgileriYazdir(self)
        print("{} dersi".format(departman))
    
    def ogrenciGecmeKalmaBelirleme(self):
        toplam=0
        ortalama=int()
        x=0
        for i in Ogrenci.ogrenciListesi:
            if i.ders==self.departman:
                toplam+=i.ogrenciNot
                x+=1
            else:
                continue
        ortalama=toplam/x
        
        for j in Ogrenci.ogrenciListesi:
            if i.ders==self.departman: 
                if j.ogrenciNot>=ortalama:
                    j.ogrenciGecmeKalma=True
                else:
                    j.ogrenciGecmeKalma=False
            else:
                continue
               
    def ogrenciBilgileriniYazdir(self):
        for i in Ogrenci.ogrenciListesi:
            if i.ders==self.departman:
                print("isim: {} soyisim:{}".format(i.isim,i.soyisim))
                print("ogrenci no: {} ogrenci notu: {}".format(i.ogrenciNo,i.ogrenciNot))
        
                if i.ogrenciGecmeKalma==True:
                    print("{} dersi basarili..\n".format(self.departman))
                else:
                    print("{} dersi basarisiz..\n".format(self.departman))
            
            else:
                continue
```

```python
ogrenci1=Ogrenci("Bahadir","AKMAN",8,40,"PC INTRO")
ogrenci2=Ogrenci("Beyzanur","AKMAN",15,90,"PC INTRO")
ogrenci3=Ogrenci("Mertcan","OZGEN",19,80,"PC INTRO")
ogrenci4=Ogrenci("Bahadir","AKMAN",8,30,"PHYSIC")
ogrenci5=Ogrenci("Beyzanur","AKMAN",15,70,"PHYSIC")
ogrenci6=Ogrenci("Mertcan","OZGEN",19,50,"PHYSIC")
ogretmen1=Ogretmen("Oguzhan","KARAHAN","PC INTRO")
ogretmen2=Ogretmen("Oktay","CEBECIOGLU","PHYSIC")
ogretmen1.ogrenciGecmeKalmaBelirleme()
ogretmen2.ogrenciGecmeKalmaBelirleme()
ogretmen1.ogrenciBilgileriniYazdir()
ogretmen2.ogrenciBilgileriniYazdir()
```
